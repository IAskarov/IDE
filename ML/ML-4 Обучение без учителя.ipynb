```python
# импортируем библиотеки numpy и sklearn
import numpy as np
from sklearn.mixture import GaussianMixture
import pandas as pd
import seaborn as sns
from sklearn.cluster import DBSCAN
```

```python
hous = pd.read_csv('housing.csv')
hous.head()
```

```python
gaussian_clustering = GaussianMixture(n_components=3)
predictions = gaussian_clustering.fit_predict(X=hous[['year', 'price']])
sns.scatterplot(x=hous.year, y=hous.price, c=predictions)
```

```python
#запускаем кластеризацию на наших данных
clustering = DBSCAN(eps=1.15, min_samples=2).fit(hous[['year', 'price']])

# Далее можно визуализировать результаты, как мы делали это с алгоритмом k-means
sns.scatterplot(x=hous.year, y=hous.price, c=clustering.labels_)
```
